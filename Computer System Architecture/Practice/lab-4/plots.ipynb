```python
import pandas as pd
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc("font", size=18)
```

```python
data = pd.read_csv("res.csv")
data.head(15)
```

```python
plt.rcParams["figure.figsize"] = (20, 10)

plt.plot(data["size"][:5], data["time"][:5], c='purple', marker='x')
plt.plot(data["size"][5:10], data["time"][5:10], c='blue', marker='.')
plt.plot(data["size"][35:40], data["time"][35:40], c='red', marker='P')
plt.grid()
plt.title('The dependence of the execution time on the volume of input data')
plt.legend(['dgemm', 'dgemm_opt_1', 'dgemm_opt_2 (BS = 64)'])
# plt.yticks(x)
plt.xticks([i for i in range(1024,6000,1024)])
plt.ylabel("Execution time")
plt.xlabel("Input volume")
plt.show()
```

```python
speedup_data1 = np.array([data["time"][:5][i] / data["time"][5:10][i + 5] for i in range(5)])
speedup_data2 = np.array([data["time"][:5][i] / data["time"][10:15][i + 10] for i in range(5)])
speedup_data3 = np.array([data["time"][:5][i] / data["time"][15:20][i + 15] for i in range(5)])
speedup_data4 = np.array([data["time"][:5][i] / data["time"][20:25][i + 20] for i in range(5)])
speedup_data5 = np.array([data["time"][:5][i] / data["time"][25:30][i + 25] for i in range(5)])
speedup_data6 = np.array([data["time"][:5][i] / data["time"][30:35][i + 30] for i in range(5)])
speedup_data7 = np.array([data["time"][:5][i] / data["time"][35:40][i + 35] for i in range(5)])

plt.plot(data["size"][5:10], speedup_data1, c='blue', marker='.')
plt.plot(data["size"][10:15], speedup_data2, c='red', marker='P')
plt.plot(data["size"][15:20], speedup_data3, c='purple', marker='P')
plt.plot(data["size"][20:25], speedup_data4, c='green', marker='P')
plt.plot(data["size"][25:30], speedup_data5, c='pink', marker='P')
plt.plot(data["size"][30:35], speedup_data6, c='gray', marker='P')
plt.plot(data["size"][35:40], speedup_data7, c='cyan', marker='P')
plt.grid()
plt.title('The dependence of the execution time on the volume of input data')
plt.legend(['dgemm_opt_1', 'dgemm_opt_2 (BS = 2)', 'dgemm_opt_2 (BS = 4)', 'dgemm_opt_2 (BS = 8)', 'dgemm_opt_2 (BS = 16)', 'dgemm_opt_2 (BS = 32)', 'dgemm_opt_2 (BS = 64)'])
# plt.yticks(x)
# plt.xticks([i for i in range(1000, 6000, 1000)])
plt.ylabel("Speedup")
plt.xlabel("Input volume")
plt.show()
```
