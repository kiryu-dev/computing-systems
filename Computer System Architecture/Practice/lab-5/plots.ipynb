```python
import pandas as pd
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc("font", size=18)
plt.rcParams["figure.figsize"] = (20, 10)
```

```python
data = pd.read_csv("res.csv")
data.head(15)
```

```python
# speedup1 = np.array([data["time"][40] / data["time"][:4][i] for i in range(4)])
# speedup2 = np.array([data["time"][40] / data["time"][20:24][i + 20] for i in range(4)])
x = np.array([int(i) for i in range(2, 9, 2)])
```

```python
speedup1 = np.array([data["time"][40] / data["time"][:4][i] for i in range(4)])
speedup2 = np.array([data["time"][40] / data["time"][20:24][i + 20] for i in range(4)])
plt.plot(x, x, c='red', marker='P')
plt.plot(x, speedup1, c='purple', marker='x')
plt.plot(x, speedup2, c='blue', marker='.')
plt.grid()
plt.title('Matrix consisting of 1000 x 1000 elements')
plt.legend(['Linear speedup', 'OpenMP', 'Posix Threads'])
plt.yticks(x)
plt.xticks(x)
plt.ylabel("Speedup")
plt.xlabel("Number of threads")
plt.show()
```
