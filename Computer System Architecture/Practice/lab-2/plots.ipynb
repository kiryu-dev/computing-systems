```python
dataset = pd.read_csv("perf.csv")
ds0 = dataset["TaskPerf"][:20]
ds1 = dataset["TaskPerf"][20:40]
ds2 = dataset["TaskPerf"][40:60]
ds3 = dataset["TaskPerf"][60:]
x = dataset["LNum"][:20]
```

```python
plt.figure().set_figheight(10)
plt.figure().set_figwidth(18)

plt.plot(x, ds0, c='purple')
plt.plot(x, ds1, c='red')
plt.plot(x, ds2, c='blue')
plt.plot(x, ds3, c='green')
plt.grid()
plt.title('double sin')
plt.legend(['O0', 'O1', 'O2', 'O3'])
plt.show()
```
