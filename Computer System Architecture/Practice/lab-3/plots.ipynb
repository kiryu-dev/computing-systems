```python
import pandas as pd
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc("font", size=18)
```

```python
data = pd.read_csv("result.csv")
data.head(15)
```

```python
from turtle import color


underplot_names = [f"{4 * i + 4} Mb" for i in range(20)]

# plt.figure().set_figheight(10)
# plt.figure().set_figwidth(18)

plt.rcParams["figure.figsize"] = (25, 10)

x = np.arange(len(underplot_names))
width = 0.3

fig, ax = plt.subplots()

rects1 = ax.bar(x - width/2, data["WriteBandwidth"][80:100], width, label='write bandwidth', color='blue')
rects2 = ax.bar(x + width/2, data["ReadBandwidth"][80:100], width, label='read bandwidth', color='green')
ax.set_title('int SSD')
ax.set_xticks(x)
ax.set_xticklabels(underplot_names)
ax.legend()
```

```python
from turtle import color


underplot_names = [f"{4 * i + 4} Mb" for i in range(20)]

# plt.figure().set_figheight(10)
# plt.figure().set_figwidth(18)

plt.rcParams["figure.figsize"] = (25, 10)

x = np.arange(len(underplot_names))
width = 0.3

fig, ax = plt.subplots()

rects1 = ax.bar(x - width/2, data["WriteBandwidth"][100:], width, label='write bandwidth', color='red')
rects2 = ax.bar(x + width/2, data["ReadBandwidth"][100:], width, label='read bandwidth', color='purple')
ax.set_title('double SSD')
ax.set_xticks(x)
ax.set_xticklabels(underplot_names)
ax.legend()
```

```python
plt.figure().set_figheight(10)
plt.figure().set_figwidth(18)

plt.plot(data["BlockSize"][80:100], data["AbsError(write)"][80:100], c='purple')
plt.plot(data["BlockSize"][80:100], data["AbsError(read)"][80:100], c='red')
plt.plot(data["BlockSize"][100:], data["AbsError(write)"][100:], c='blue')
plt.plot(data["BlockSize"][100:], data["AbsError(read)"][100:], c='green')
plt.grid()
plt.title('            ')
plt.legend(['int write', 'int read', 'double write', 'double read'])
plt.show()
```

```python
plt.figure().set_figheight(10)
plt.figure().set_figwidth(18)

data_write_ram = [62814.1, 35919.5, 42936.9, 36462.3]
data_read_ram = [34059.9, 13287.3, 15798.6, 11266.4]
x = [131072, 1048576, 6291456, 16777216]
xnames = ['128Kb', '1Mb', '6Mb', '16Mb']
plt.plot(x, data_write_ram, c='purple', marker='.')
plt.plot(x, data_read_ram, c='red', marker='*')
plt.grid()
plt.xticks(x, xnames)
plt.title('          RAM')
plt.legend(['write', 'read'])
plt.ylabel("Bandwidth (Mb/s)")
plt.show()
```

```python
x = [4 * 1024 * 1024 + i * (4 * 1024 * 1024) for i in range(20)]
xnames = [f"{4 * i + 4} Mb" for i in range(20)]
plt.plot(x, data[40:60]["WriteBandwidth"], c='blue', marker='.')
plt.plot(x, data[40:60]["ReadBandwidth"], c='red', marker='*')
plt.grid()
plt.xticks(x, xnames)
plt.title('          SSD')
plt.legend(['write', 'read'])
plt.ylabel("Bandwidth (Mb/s)")
plt.show()
```

```python
x = [4 * 1024 * 1024 + i * (4 * 1024 * 1024) for i in range(20)]
xnames = [f"{4 * i + 4} Mb" for i in range(20)]
plt.plot(x, data[40:60]["RelError(write)"], c='purple', marker='.')
plt.plot(x, data[60:]["RelError(write)"], c='blue', marker='*')
plt.grid()
plt.xticks(x, xnames)
plt.title('          SSD  flash')
plt.legend(['SSD', 'flash'])
plt.ylabel("Relative error write (%)")
plt.show()
```
