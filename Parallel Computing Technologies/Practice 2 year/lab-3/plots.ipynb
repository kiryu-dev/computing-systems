```python
import pandas as pd
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc("font", size=18)
```

```python
data = pd.read_csv("data.csv")
data.head(4)
```

```python
x = data["Linear"]
x_names = ["24\n3/8", "32\n4/8", "40\n5/8", "48\n6/8", "56\n7/8"]
```

```python
plt.rcParams["figure.figsize"] = (20, 10)

plt.plot(x, data["Linear"], c='purple', marker='x')
plt.plot(x, data["SGEMV1"], c='blue', marker='.')
plt.plot(x, data["SGEMV2"], c='red', marker='P')
plt.grid()
plt.title('Parallel SGEMV')
plt.legend(['Linear speedup', 'SGEMV (n = 28000)', 'SGEMV (n = 45000)'])
plt.yticks(x)
plt.xticks(x, x_names)
plt.ylabel("Relative speedup")
plt.xlabel("Number of processes")
plt.show()
```
